```python
#get all links 
anchor=soup.find_all("a")
alllinks=set()
for link in anchor:
    if link!="#":
        flink= url +link.get("href")
        alllinks.add(link)
        print(flink)
```

```python
#to get topic titles,first we will create a variable to store the parsed contnt (i.e soup)
topic_titles=soup.find_all("p",{"class":"f3 lh-condensed mb-0 mt-1 Link--primary"})
topic_titles
```

```python
topic_description=soup.find_all("p",{'class':'f5 color-fg-muted mb-0 mt-1'})
topic_description[:6]
```

```python
all_topic_titles=[]
for topic in topic_titles:
    all_topic_titles.append(topic.text)
    
print(all_topic_titles)
```

```python
#same for topic descriptions
all_top_des=[]
for tags in topic_description:
    all_top_des.append(tags.text.strip())
print(all_top_des)
```

```python
topic_link=soup.find_all('a',{"class": "no-underline flex-grow-0"})
first_link=url.split("/t")[0]+topic_link[0]['href']
print(first_link)
```

```python
all_links=[]
base_url="https://github.com"
for link in topic_link:
    all_links.append(base_url+link["href"])

print(all_links)
```

### create a csv file to extract data within it

```python
import pandas as pd
```

```python
data={"topic_title":all_topic_titles
      ,"description":all_top_des,
      "url":all_links}
topics=pd.DataFrame(data)
topics
```

```python
topics.to_csv("topics.csv",index=None)
```

```python
def star_count(stars_str):
    stars_str=stars_str.strip()
    if stars_str[-1]=="k":
        return int(float(stars_str[:-1])*1000)
    return int(stars_str)
```

```python
star_count(star_tags[0].text.strip())
```

```python
def get_repo_info(h1_tags,star_tag):
    '''give all information about repository'''
    
    a_tags=h1_tags.find_all("a")
    username = a_tags[0].text.strip()
    repository_name = a_tags[1].text.strip()
    repository_url = base_url+a_tags[1]["href"]
    star= star_count(star_tag.text.strip())
    return username,repository_name,repository_url,star
```
